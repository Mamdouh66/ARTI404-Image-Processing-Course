### STEP 4: Reading an image

```python
from PIL import Image
import sys
import numpy as np

try:
    img = Image.open("MyMemoji.png")

except IOError:
    print("Unable to load image")
    sys.exit(1)

# Opens a window and displays the image
img.show()

# Prints the image into ipynb
from IPython.display import Image as IPImage

IPImage(filename="MyMemoji.png")
```

### Getting Basic information from images

```python
print("Format: ", img.format, "\nSize: ", img.size, "\nMode: ", img.mode)
```

### Step 5: Perform some modifications on an image 

#### 1- Rotating an image

```python
rot = img.rotate(180)
rot.save("MyMemoji_rotated.png")
rot.show()  # Opens a window and displays the image
IPImage(filename="MyMemoji_rotated.png")
```

#### 2-	Rotate Image by using Transpose function

```python
from PIL.Image import Transpose

rot3 = img.transpose(Transpose.ROTATE_90)
rot2 = img.transpose(Transpose.ROTATE_270)
rot3.save("MyMemoji_rotated3.png")
rot2.save("MyMemoji_rotated2.png")
rot2.show()
rot3.show()

IPImage(filename="MyMemoji_rotated2.png")
```

```python
IPImage(filename="MyMemoji_rotated3.png")
```

#### 3- Crop Image

```python
crop = img.crop((100, 100, 150, 150))
crop.show()

crop.save("MyMemoji_cropped.png")
IPImage(filename="MyMemoji_cropped.png")
```

#### 4-	Copy a subrectangle from an image

```python
box = (100, 100, 400, 400)
region = img.crop(box)
region.show()

region.save("MyMemoji_region.png")
IPImage(filename="MyMemoji_region.png")
```

## Task 1: Read the image Mario and display it like this. 


```python
from PIL import Image

try:
    img = Image.open("Mario.png")

except IOError:
    print("Unable to load image")
    sys.exit(1)
```

```python
flipped_img = img.transpose(Image.FLIP_TOP_BOTTOM)
turned_img = flipped_img.transpose(Image.FLIP_LEFT_RIGHT)

combined_img = Image.new("RGB", (img.width * 2, img.height))

combined_img.paste(img, (0, 0))

combined_img.paste(turned_img, (img.width, 0))

combined_img.save("Mario_flipped.png")

IPImage(filename="Mario_flipped.png")
```

## Task 2: Convert the same image into 2nd display image.

```python
flipped_img = img.transpose(Image.FLIP_TOP_BOTTOM)
turned_img = flipped_img.transpose(Image.FLIP_LEFT_RIGHT)

combined_img = Image.new("RGB", (img.width * 2, img.height))

combined_img.paste(img, (0, 0))

combined_img.paste(turned_img, (img.width, 0))

combined_img.save("2nd_display_image.png")

IPImage(filename="2nd_display_image.png")
```

### Task 3: Convert the Mario image into Gray Scale. 

```python
gray_img = img.convert("L")
gray_img.save("Mario_gray.png")
# gray_img.show()

IPImage(filename="Mario_gray.png")
```
